## droping the unused columnsÂ¶


```python
def get_unused_cols(x,y):
    # getting the names of the columns i won't use in my analysis
    z = []
    for e in y:
        if e not in x:
            z.append(e)
    return z
```

```python
# geting the names of the columns will be removed
needed_col = ['ListingKey', 'Term', 'LoanStatus', 'BorrowerRate', 'EstimatedLoss', 'EstimatedReturn','ProsperRating (Alpha)', 'ProsperScore', 'ListingCategory (numeric)', 'EmploymentStatus', 'CreditScoreRangeLower', 'CreditScoreRangeUpper', 'DebtToIncomeRatio', 'IncomeRange', 'TotalProsperLoans', 'IsBorrowerHomeowner', 'MonthlyLoanPayment','LoanOriginalAmount', 'LoanOriginationQuarter','Investors']
df_col_name = df.columns
# storing the columns that will be droped in remove_cols list
remove_cols = get_unused_cols(needed_col,df_col_name)
len(remove_cols)
```

```python
# drop the columns 
df.drop(remove_cols,axis=1, inplace=True)
```

```python
# compute the logarithm of EstimatedLoss to make multivariate plotting easier
def log_trans(x, inverse = False):
    """ quick function for computing log and power operations """
    if not inverse:
        return np.log10(x)
    else:
        return np.power(10, x)

df['log_EstimatedLoss'] = df['EstimatedLoss'].apply(log_trans)
```

```python
# compute the cuberoot of BorrowerRate to make multivariate plotting easier
def cuberoot_trans(x, inverse = False):
    """ quick function for computing cube root and cube operations """
    if not inverse:
        return x ** (1/3)
    else:
        return x ** 3
df['cr_borrower_rate'] = df['BorrowerRate'].apply(cuberoot_trans)
```
